```python
from sqlalchemy import create_engine
import pandas as pd
from gensim.corpora.dictionary import Dictionary
from gensim.models.tfidfmodel import TfidfModel
from gensim.matutils import corpus2csc
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def sql_to_list_converter(chunks):
    preprocessed_texts = []
    for i in chunks:
        
        #have to split each sentence since this is the input for gensim
        list_to_add = [iterator.split() for iterator in i.preprocessed_text.tolist()]
        preprocessed_texts += list_to_add
        
    return preprocessed_texts

def tf_idf_vect(preprocessed_texts):
    dictionary = Dictionary(preprocessed_texts)
    num_docs = dictionary.num_docs
    num_terms = len(dictionary.keys())
    
    #transform into bow
    corpus_bow = [dictionary.doc2bow(doc) for doc in preprocessed_texts]
    
    #transform into tf-idf:
    tfidf = TfidfModel(corpus_bow)
    corpus_tfidf = tfidf[corpus_bow]
    
    #transform into sparse matrix:
    corpus_tfidf_sparse = corpus2csc(corpus_tfidf, num_terms,num_docs=num_docs)
    return corpus_tfidf_sparse

def log_scaler(num, mean, std):
        #num_log = np.log(num)
        num_log_scaled = (num - mean)/std
        
        return num_log_scaled

def scaler_lists(my_list):
    #log_list = [np.log(n) for n in my_list]
    std = pd.Series(my_list).std()
    mean = pd.Series(my_list).mean()

    scaled_list = [log_scaler(i, mean, std) for i in my_list]
    return scaled_list
```

```python
def chunks_to_lists(chunks):
    
    favorite_count = []
    retweet_count = []
    reply_count = []
    quote_count = []
    
    for i in chunks:
        favorite_count.append(list(i.favorite_count))
        retweet_count.append(list(i.retweet_count))
        reply_count.append(list(i.reply_count))
        quote_count.append(list(i.quote_count))
        
    # flatten out lists:
    favorite_count = [item for sublist in favorite_count for item in sublist]
    retweet_count = [item for sublist in retweet_count for item in sublist]
    reply_count = [item for sublist in reply_count for item in sublist]
    quote_count = [item for sublist in quote_count for item in sublist]

    
    return favorite_count, retweet_count, reply_count, quote_count
```

```python
favorite_count, retweet_count, reply_count, quote_count = chunks_to_lists(chunks)
```

```python
favorite_count_scaled = scaler_lists(favorite_count)
retweet_count_scaled = scaler_lists(retweet_count)
#reply_count_scaled = scaler_lists(reply_count)
#quote_count_scaled = scaler_lists(quote_count)
```
